## __벡터 및 행렬 표기법__

임의의 열벡터 $v =(a_1,b_1,c_1)^T , \ w=(a_2,b_2,c_2)^T$는 각각 $\begin{pmatrix} a_1 \\ b_1 \\ c_1 \end{pmatrix}, \ \begin{pmatrix} a_2 \\ b_2 \\ c_2 \end{pmatrix}$ 으로 표기되며, $v$ 와 $w$ 열벡터를 행렬로 나타내면 다음과 같다.
$$ \begin{pmatrix} a_1 & a_2 \\ b_1 & b_2 \\ c_1 & c_2 \end{pmatrix}$$

* __선형 결합: $\alpha v + \beta w$__  
$$ \Rightarrow \begin{pmatrix} a_1 & a_2 \\ b_1 & b_2 \\ c_1 & c_2 \end{pmatrix} \begin{bmatrix} \alpha \\ \beta \end{bmatrix} = \alpha \begin{bmatrix} a_1 \\ b_1 \\ c_1  \end{bmatrix} + \beta\begin{bmatrix} a_2 \\ b_2 \\ c_2  \end{bmatrix}$$  


## __열공간을 기반한 해석__

###  $\circledcirc$ 전진 소거법 

#### __Example__

$$
2u + v + w =5 \\ 4u - 6v = -2 \\ -2u + 7v + 2w = 9 
$$

이때, 1 행의 $u$의 계수 2가 첫 번째 피벗이 된다. __: 단, 피벗은 0이 아닌 정수__

1 행의 계수를 통해 2 행과 3행에 $u$의 계수의 배수 만큼 연산.   

$$
2u + v + w =5 \\ - 8v  - 2w = -12 \\ 8v + 3w = 14 
$$  

__: 두 번째 피벗은 $v$의 계수 -8__  

$w, v$ 도 계산하면, 행렬은 상삼각 행렬이 가지게 되며 각 변수는 유일한 해를 얻게 된다.

$$
2u + v + w =5 \\ \quad \quad - 8v  - 2w = -12 \\ \quad \quad \qquad w = 2 
$$


$\rightarrow \ w=2, \rightarrow \ v=1, \rightarrow \ u=1\ \Rightarrow$ back-substitution  



### $\circledcirc$ 소거법 계산 시 주의할 점  

피벗이 0일 경우, 소거법을 중단하고 연립 방정식의 순서를 바꿔야 한다. $\Rightarrow$ Pivotting(피보팅)

#### __Example1__  

$ u+v+w = a \\ 2u+2v+5w = b \\ 4u+6v+8w = c$ $\quad \rightarrow \quad$ $ u+v+w = a \\ \qquad \quad 3w = b-2a \\ \quad 2v+4w = c-4a$ $\quad \rightarrow \quad$ __Pivotting__ $\quad \rightarrow \quad$ $u+v+w = a \\ \quad \ 2v+4w = c-4a \\ \qquad \quad 3w = b-2a$

#### __Example2__  

$ u+v+w = a \\ 2u+2v+5w = b \\ 4u+4v+8w = c$ $\quad \rightarrow \quad$ $ u+v+w = a \\ \qquad \quad 3w = b-2a \\ \quad \qquad 4w = c-4a$  

* 2행의 피벗이 0이므로, 피보팅을 할 수 없다.  

__$\frac{b-2a}{3} = \frac{c-4a}{4}$ 식을 만족하는 해가 무수히 많다. 그렇지 않으면 해가 없다.__
