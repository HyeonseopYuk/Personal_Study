## ◎ __행렬 A의 열 공간(Column Space)__  

○ __Column Space__ : 행렬 A의 열벡터로 구성된 모든 선형 결합(기저)  

$$Ax = b \Rightarrow \begin{bmatrix} 1 & 1 & \cdots & 1 \\ a_1 & a_2 & \cdots & a_n \\ \vdots & \vdots & \ddots & \vdots \\ 1 & 1 & 1 & 1 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix} = \begin{bmatrix} b_1 \\ b_2 \\ \vdots \\ b_n \end{bmatrix} $$

* $x_1a_1 + x_2a_2 + \cdots + x_na_n = b$ : 열 벡터의 선형 결합  

## ○ __위 식에서 $b$가 Column space 속할 경우 해가 존재한다.__

__Example__  

$$
u \begin{bmatrix} 1 \\ 5 \\ 4 \end{bmatrix} + v \begin{bmatrix} 0 \\ 4 \\ 4 \end{bmatrix} = \begin{bmatrix} b_1 \\ b_2 \\ b_3 \end{bmatrix}
$$

__※ $(b_1, b_2, b_3)$: (1,5,4) & (0,4,4) 공간 상에 놓여 있음.__

## ◎ __Column space $\quad \mathcal{C}(A)$__  

1) (가정) $\ b, b' \in \mathcal{C}(A)$일 경우, $\rightarrow A(x+x') = b+b'$  
: $Ax_1 =b_1 , \quad Ax_2 =b_2, \quad b_1+b_2 = b \\ A(x_1)+A(x_2) = A(x_1+x_2) = b$ 

2) (가정) $\ b \in \mathcal{C}(A)$일 경우 $\rightarrow A(cx)= cb$

## ◎ __Null Space of A $\quad \mathcal{N}(A)$__  
$$\mathcal{N}(A) = \{x|Ax=0\}$$  

$Ax=0$ __방정식이 모수 보다 더 많은 경우(즉, m>n) : 무수히 많은 해 존재__  
$\rightarrow$ 행렬 A의 null Space - Vector Space 상에서 존재.

○ 행렬 A의 null space는 x의 모든 벡터로 구성되어 있다. $Ax=0$을 만족하는 $\mathcal{N}(A) \rightarrow$ subspace  

1) $Ax=0,\  Ax'=0 \quad \rightarrow \quad A(x+x')=0,$ __$\quad x_1+x_2 \in \mathcal{N}(A) $__  
2) $Ax=0 \quad \rightarrow Acx =c \cdot 0 = 0, \quad cx \in \mathcal{N}(A)$

__Example__  

$$(1)\quad  A = \begin{bmatrix} 1 & 0 \\ 5 & 4 \\ 2 & 4  \end{bmatrix} \begin{bmatrix} u \\ v \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix} \quad \rightarrow \quad (0,0) \in \mathcal{N}(A) \\ (2) \quad  B = \begin{bmatrix} 1 & 0 & 1 \\ 5 & 4 & 9  \\ 2 & 4 & 6  \end{bmatrix} \begin{bmatrix} u \\ v \\ w\end{bmatrix} = \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix} \rightarrow \mathcal{C}(A) = \mathcal{C}(B) $$
$B$에 대해 가우스 소거법으로 풀면 $(c,c,-c)$ = (1,1,-1)이 성립 되므로, $\begin{bmatrix} u \\ v \\ w \end{bmatrix} = c\begin{bmatrix} 1 \\ 1 \\ -1 \end{bmatrix}$ 과 같다.

▒ __$\mathcal{N}(B)$ : 원점을 지나는 직선상에서 나타나며 무수히 많은 해를 지님.__

## 2.6 Linear Transformation

#### Ax=b

: 위 식을 다르게 접근해보자!
* x(입력 벡터) -> A (System equation) -> b(출력 벡터)

A: 선형변환 ($\mathbb{R}^n \ \rightarrow \ \mathbb{R}^m$)  
ex) 상수배 확장 및 수축, 회전, 대칭, 정사영

### __선형성이 보장되는 모든 연산, 수식에 대해서는 행렬 식으로 변환할 수 있다.__
ex) __differentaiation__: $\frac{d}{dt}$ 은 $\mathbb{R}^{n+1} \rightarrow \mathbb{R}^n$이 된다. 

$X(t) = a_0+a_1t+a_2t^2+\cdots+a_nt^n \quad \Rightarrow \quad X = \begin{bmatrix} a_0 \\ a_1 \\ \vdots \\ a_n \end{bmatrix} \rightarrow_{\frac{d}{dt}} \quad y=\begin{bmatrix} b_0 \\ b_1 \\ \vdots \\ b_{n-1} \end{bmatrix}$

ex) __integration__: $\int_{-\infty}^{t} a_0+a_1t+\cdots + a_nt^n\ dt$ 은 $\mathbb{R}^{n} \rightarrow \mathbb{R}^{n+1}$이 된다. 

$X(t) = a_0t+\cdots+\frac{a_n}{n+1}t^{n+1} \quad \Rightarrow \quad X = \begin{bmatrix} a_0 \\ a_1 \\ \vdots \\ a_n \end{bmatrix} \rightarrow \quad y=\begin{bmatrix} 0 \\ b_1 \\ \vdots \\ b_{n-1} \end{bmatrix}$

ex) __Multiplication__  (2차식이라 가정) : $\mathbb{R}^{n} \rightarrow \mathbb{R}^{n+2}$
$x(t)p(t) = y(t)\ \rightarrow \ (a_1x_1(t)+a_2x_2(t))p(t) = a_1x_1(t)p(t)+a_2x_2(t)p(t) $ 
$$
\Rightarrow p(t) = p_0+p_1t+p_2t^2이므로, \quad \; x(t)p(t) = b_0+b_1t + \cdots + b_{n+2}t^{n+2}
$$
