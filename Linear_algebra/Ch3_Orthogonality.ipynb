## 3.1 직교 벡터와 부분 공간

○ 직교성 : 서로 독립인 기저  

○ 데카르트 좌표계(2차원 좌표계)  

○ 벡터의 길이 : $|| x ||^2$  

○ 각 축이 직각을 만족할 경우 피타고라스 정리에 의해 계산할 수 있다.

$$|| x ||^2 + || y ||^2 = || x-y ||^2 \\ \Rightarrow x^Tx + y^Ty = (x-y)^T(x-y)(:x^Tx-x^Ty-y^Tx-y^Ty) \\ ※ x^Ty = y^Tx = 0 \ for \ perpendicular \\ \Rightarrow x^Ty : inner \ product$$ 

◎ $x^Ty = 0 \quad :orthogonal \\ x^y<0 \quad : angle > 90^{o} \\ x^y>0 \quad : angle < 90^{o}$  

◎ 0이 아닌 벡터들은 서로 직교이다 $\rightarrow$ 벡터 간 독립임을 의미  
pf) $c_1v_1+c_2v_2+\cdots+c_nv_n=0$이 있다고 할 때, $v_i^T(c_1v_1+c_2v_2+\cdots+c_nv_n)=0 \\ \Rightarrow c_i||v_i||^2=0. \ \forall i \\ \Rightarrow c_i=0$  

만약, $||v_i||=1$일 경우, orthogonal normal이 된다.


◎ __orthogonal subspaces__     
$\Rightarrow$ 원점을 지나는 벡터로부터 임의의 부분 공간을 포괄(span)하는 기저 벡터들의 선형결합이 수직인 공간을 의미한다.

$v \in V, \quad w \in W \qquad v^Tw=0 \; \forall \ v \ \& \ w$

◎ Row space $\bot$ Null space $\in \mathbb{R}^n$   
◎ Columns space $\bot$ left null space $\in \mathbb{R}^m$  
 
◎ 직교 여공간: 주어진 부분 공간과 수직인 벡터 공간  

$v \in V, \quad w \in W \quad \in \mathbb{R}^n$  

$v \bot w \rightarrow $ 직교 부분공간  
Dim(V) + Dim(W) = n

◎ 직교 여공간 $\neq$ 직교 부분공간

## 3.3 정사영과 최소제곱법

* 정방 행렬인 경우 가우스 소거법으로 해를 풀 수 있다.

## 3.4  직교 기저  

○ 벡터가 직교할 경우 $\rightarrow$ 벡터 간 독립이다. 이는 벡터 공간을 포괄(span)할 수 있다는 의미이다.  

○ $q_1,q_2, \cdots, q_n$이 서로 수직이며 크기가 1이라고 하자. 이를, 정규직교기저라고 한다. $$q_i^Tq_j = \begin{cases} 0, \quad i \neq j \\ 1, \quad i=j \end{cases}, \qquad \qquad Q=\begin{bmatrix} q_1,q_2, \cdots, q_n \end{bmatrix}$$
$Q^TQ \Rightarrow Q^T=Q^{-1}$가 되므로 $Q^T$를 left inverse이다.  

◎ __Q example__  

1) 회전 행렬 : $\begin{bmatrix} cos \theta & -sin \theta \\ cos \theta & cos \theta \end{bmatrix}$  

2) 치환 행렬 : $\begin{bmatrix} 0 & 1 & 0 \\ 1 & 0 & 0 \\ 0 & 0 & 1 \end{bmatrix}$  

◎ __Q 변환은 크기와 회전을 유지하는 특징을 가지고 있다.__  

$||X||^2 = X^TX, \quad ||QX||^2 = X^TQ^TQX = X^TX$  

$X^Ty, \quad (QX)^T(Qy) = X^TQ^TQy = X^Ty$  

○ __정사영은 길이가 줄어드는 특징이 있다.__

$||PX|| \leq ||X||$  

○ 다음과 같이 벡터 간 서로 독립이고 직교하는 벡터 $q_1,q_2, \cdots, q_n \in \mathbb{R}^n$가 존재한다고 할 때, 어떠한 벡터공간이더라도 span 할 수 있다.(__정방 행렬__)  

$X = \sum_{i=1}^n c_iq_i,\; X=\begin{bmatrix} q_1,\cdots, q_n \end{bmatrix} \begin{bmatrix} c_1 \\ c_2 \\ \vdots \\ c_n \end{bmatrix}$ = __Q의 역행렬 통해 C를 구할 수 있다.__  

(행렬로 풀이) $c_i = \frac{q_i^TX}{q_i^Tq_i}$ : X 벡터가 $q_i$ 방향에 정사영하였을 때 크기 값이다.  

(벡터 연산) $q_i^TX = q_i^T(\sum_{j=1}^n c_jq_j)=q_i^T(c_1q_1+\cdots+c_nq_n) = c_i||q_i||^2 = c_i$  

※ $X=\sum_{i=1}^n (q_i^TX)q_i$  


○ 다음과 같이 벡터 간 서로 독립이고 직교하는 벡터 $q_1,q_2, \cdots, q_n \in \mathbb{R}^n$가 존재한다고 할 때, 어떠한 벡터공간이더라도 span 할 수 있다.(__직사각형 행렬__)

$\qquad \qquad \qquad \qquad \qquad \qquad Q=\begin{bmatrix} q_1,\cdots, q_m \end{bmatrix}, \quad m < n$__인 경우만 존재__.  


Qx=b이 성립할 때, $X=(Q^TQ)^{-1}Q^Tb)=Q^Tb$이 만족한다. 

image.png

$\qquad(n\times m) \quad \quad \qquad \qquad (m \times n) \qquad \qquad \qquad \qquad \qquad  (n \times n)$  

__example__

$Q = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}, Q^T = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \end{bmatrix}$일때,  


$\Rightarrow Q^TQ = \begin{bmatrix} 1&0 \\ 0 &1 \end{bmatrix}, QQ^T = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 0 \end{bmatrix}$이므로, Q는 left inverse일 경우에만 존재한다.  


#### ○ __A=QR factorization__  

$\begin{bmatrix} a_1 & a_2 & \cdots & a_n \end{bmatrix}=\begin{bmatrix} (q_1^Ta_1)q_1 & (q_1^Ta_2)q_1+(q_2^Ta_2)q_2 & \cdots & (q_1^Ta_n)q_1+(q_2^Ta_n)q_2 + \cdots + (q_n^Ta_n)q_n \end{bmatrix}$  

$\Rightarrow$ 각 열벡터의 선형결합으로 표현하면 다음과 같다.  

$\begin{bmatrix} q_1 & \cdots & q_n \end{bmatrix}=\begin{bmatrix} (q_1^Ta_1) & (q_1^Ta_2) & \cdots & (q_1^Ta_n) \\ 0 & (q_2^Ta_2) & \cdots & (q_2^Ta_n) \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \cdots & (q_n^Ta_n) \end{bmatrix}$ = __QR__ 이 되며 R은 상삼각행렬이다.   

$AX=b$일때, $X = (A^TA)^{T}b = (R^TQ^TQR)^{-1}R^TQ^Tb = R^TRR^TQ^Tb$이 된다.  


## ○ 함수 공간 및 퓨리에 변환

__벡터 공간 $(\mathbb{R}^n)$ 과 힐버트 공간$(\mathcal{R}^{\infty})$의 비교__  

* 원소 : 벡터 ($v_1,v_2,\cdots$) :: 함수 ($x_1(t),x_2(t),\cdots$)  

* 독립성 : 기저 벡터 간 독립이므로 선형 결합 $\rightarrow$ :: 기저 함수의 $t$시점에 대해 독립성을 가진다.

* $X= \sum_{i=1}^n c_i q_i \rightarrow$ 내적을 통해 직교 벡터 표현 가능  ::            $X= \sum_{t=1}^{\infty} a_i b_i(t) \rightarrow $ 함수간 내적을 통해 직교 함수 표현 가능(다항 함수)

##### 함수 공간(힐버트 공간)    


$\Rightarrow$ 유한 차원의 벡터 공간을 무한 차원의 함수공간으로 확장 시킨 개념  

$\Rightarrow \ x(t), y(t) \in H, \quad a\leq t \leq b$  

$\Rightarrow \ x(t)+y(t) \in H, \quad a\leq t \leq b$  

$\Rightarrow \ \alpha x(t) \in H, \quad a\leq t \leq b$  

즉, 함수 공간을 벡터처럼 표현해주기 위한 것으로, 아래와 같이 나타낼 수 있다.  

$\lim_{\Delta t \rightarrow 0 } x(t) = (x(a), x(a+\Delta t), x(a+\Delta 2t), \cdots ,x(b))^T \rightarrow x(t) \in \mathbb{R}^{\infty}$  

벡터와 내적연산과 함수의 내적 연산을 비교하면 다음과 같다.  

$x^Ty = \sum_{i=1}^n x_iy_i \quad \Rightarrow \quad (x(t),y(t)) = \lim_{\Delta t \rightarrow 0 } \sum_{k=0}^{\infty} x(a+k\Delta t)y(a+k\Delta t) = \int_a^b x(t)y(t)dt$  


＊ $||x(t)||^2 = \int_a^b x^2(t) dt$  


#### 직교성

$X^Ty = 0, \quad (x(t),y(t)) = \int_a^b x(t)y(t)dt=0$  

○ 기저 벡터 $\rightarrow$ 기저 함수  

: $X = \sum_{i=1}^n a_iv_i \quad \rightarrow \quad x(t) = \sum_{i=1}^{\infty} a_ib_i(t)$ (series의 형태)

: $\sum_{i=1}^n (q_i^Ta_i)q_i \quad \rightarrow \quad \sum_{i=1}^{\infty} (q_i(t),x_i(t))q_i(t)$  


○ 만약 임의의 기저 벡터(함수)가 존재할 때, 기저벡터의 선형 결합으로 만들어 내는 계수(series 계수)는 유일하다.
