# 4.2 행렬식의 특징  

○ $2 \times 2 $ 행렬의 행렬식을 살펴보고 이를 $n \times n$으로 확장시켜보자. $\quad \rightarrow det \begin {bmatrix} a & b \\ c & d \end{bmatrix} = ad-bc$  

◎ 3가지의 기본적인 특징  

1. $det(I) = 1$  

2. 서로 다른 행의 위치가 서로 바뀔 때, $det(A)$의 부호도 바뀐다.  

3. $det(A)$는 첫 번째 행에 대해서 선형 독립이다. 

__example__  

$1) \begin{vmatrix} a+a' & b+b' \\ c & d \end{vmatrix} = \begin{vmatrix} a & b \\ c & d \end{vmatrix} + \begin{vmatrix} a' & b' \\ c & d \end{vmatrix},  \quad 2) \begin{vmatrix} ta' & tb' \\ c & d \end{vmatrix} = t\begin{vmatrix} a & b \\ c & d \end{vmatrix}$  

※ 주의해야할 점  

* $det(A + B) \neq det(A) + det(B)$  

* $det(tA) \neq t\ det(A) \Rightarrow t^ndet(A)$  

◎ 위 특징을 바탕으로 추가적인 확장을 고려해볼 수 있다.  

4. 두 행이 같을 경우, $det(A) = 0$ 이 된다.  

__example__  

$A = \begin{bmatrix} a_1^T \\ a_2^T \\ a_3^T \end{bmatrix}, a_1^T=a_2^T \; \rightarrow \; A'=\begin{bmatrix} a_2^T \\ a_1^T \\ a_3^T \end{bmatrix}$이고, $det(A)=-det(A')$이 된다.  
이 때, 위 식을 만족하는 조건은 $det(A)=0$일 때이다.  

5. 행 연산을 하더라도 행렬식은 변하지 않는다.  
$\begin{vmatrix} a-lc & b-ld \\ c & d \end{vmatrix} = \begin{vmatrix} a & b \\ c & d \end{vmatrix} + \begin{vmatrix} -lc & -ld \\ c & d \end{vmatrix} = \begin{vmatrix} a & b \\ c & d \end{vmatrix}$  


6. A의 행 벡터가 0일 경우, $det(A) = 0$  


7. A가 삼각행렬일 경우, $det(A)$는 대각성분의 곱 연산이다.  


8. $\begin{cases} A: singular \rightarrow det(A) = 0 \rightarrow A^{-1} not \ exist \\ A: non-singular \rightarrow det(A) \neq 0 \rightarrow A^{-1}  exist \end{cases}$  


9. $det(AB) = det(A) det(B)$  


10. $det(A^T) = det(A)$  

행렬 A에 가우스 소거법으로 삼각행렬로 변환하는데, 이때 생기는 모든 피벗들의 곱으로 연산되는 것이 행렬 A의 행렬식이다. 식으로 표현하면 아래와 같다.

$A=LDU \rightarrow det(A) = 1\cdot det(D) \cdot 1$이 성립한다.  

$A^T= U^TD^TL^T \rightarrow det(A^T) = 1\cdot det(D) \cdot 1$이 성립한다.  

이를 일반화하면 아래와 같다.  

$PA = LDU \quad$  이때, $det(P)= \pm1, \quad det(A) = \pm(\Pi_{i=1}^n P_i)$

# 4.4 행렬식의 응용

$\Rightarrow$ 행렬식 구하는 방법  

1) 가우스 소거법 $A \rightarrow U$  
행렬식은 상삼각행렬의 대각행렬인 피벗의 곱으로 계산할 수 있다. : $\Pi_{i=1}^n P_i$  


2) Big formula  

$\sum_{\alpha, \beta, \gamma, \cdots, \mathcal{V}} (a_{1\alpha},a_{2\beta},a_{3\gamma}, \ codts, a_{n\mathcal{V}})detP_{\alpha,\beta,\gamma, \cdots, \mathcal{V}}, \quad (\alpha, \beta, \gamma, \cdots, \mathcal{V}) \rightarrow$ 순열의 조합  


3) 여인수 전개 $\rightarrow$ 특정 행에 집중 

$\sum_{j=1}^n a_{ij}c_{ij} = \sum_{j=1}^n a_{ij}(-1)^{i+j}det(M_{ij})$

○ 역행렬 계산  

$A^{-1} = \frac{\mathcal{C}^T}{det(A)} \Rightarrow (det(A))I = A\mathcal{C}^T$, $\qquad \mathcal{C} : $ 여인수 행렬  

$\begin{bmatrix} a_{11} & a_{12} & \cdots & a_{1n} \\ \vdots & \cdots & \ddots& \vdots \\ a_{n1} & & & a_{nn} \end{bmatrix} \begin{bmatrix} c_{11} & c_{12} & \cdots & c_{1n} \\ \vdots & \cdots & \ddots& \vdots \\ c_{n1} & & & c_{nn} \end{bmatrix}= \begin{bmatrix} det(A) & & & \\  & det(A) & & & \\ & & \ddots &   \end{bmatrix}$

여인수 행렬은 특정 열과 행을 제외한 행렬에서 얻어지는데, 서로 다른 순서의 행렬 A의 행벡터와 $\mathcal{C}$의 열벡터가 연산이 되는 경우 중복되어 0이 된다. 따라서 행렬식은 0이 된다.  

○ $Ax=b, \; X=\begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix} = A^{-1}b =  A^{-1}\begin{bmatrix} b_1 \\ b_2 \\ \vdots \\ b_n \end{bmatrix}$이므로, 크래머의 법칙으로 $x_i = \frac{det(B_i)}{det(A)}$ 풀 수 있다.  
※ $B_i = \begin{bmatrix} a_1 & a_2 & \cdots & a_{i-1} & b & a_{i+1} & \cdots & a_n \end{bmatrix}$  

○ $Ax=b$를 여인수 전개로 해를 푸는 방법 소개.  

$x=A^{-1}b = \frac{c^Tb}{det(A)} \Rightarrow x_i = \frac{det(B_i)}{det(A)}$  

$B_i = \begin{bmatrix} a_{11} & \cdots & b_1 & \cdots & a_{1n} \\  \vdots & \vdots & \vdots & \vdots & \vdots \\ a_{n1} & \cdots & b_n & \cdots & a_{nn}  \end{bmatrix}, \; det(B_i) = b_1c_{1i} + b_2c_{2i} + \cdots +b_nc_{ni}$이고,  $b$가 열벡터 이므로 $A^T$형태로 행렬식 구해진다.
